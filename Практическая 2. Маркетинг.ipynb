# Создадим новую категориальную переменную на основе колличественной переменной "Age"

```python
 def get_age_group_cat(age_group):
    if age_group < 18:
        return 'Младше 18'
    elif age_group < 25:
        return '18 - 25 лет'
    elif age_group < 35:
        return '26 - 35 лет'
    elif age_group < 45:
        return '36 - 45 лет'
    elif age_group < 55:
        return '46 - 55 лет'
    elif age_group >= 55:
        return 'Старше 55'
    else: 
        return age_group
```

```python
df ['Age_Group'] = df ['Age'].apply(get_age_group_cat)
```

# Проверка гипотез

#### Проверим существует ли статистически значимая корреляция между наличием маленьких детей в семье и принятием решения о покупке в первую волну компании.

```python
# Выдвигаем гипотезы:
# H_1 : Между наличием маленьких детей в семье и решением совершить покупку в первую волну компании связь есть
# H_0 : Между наличием маленьких детей в семье и решением совершить покупку в первую волну компании связи нет

#Для поиска связи будем применять непаратемтрический критерий проверки гипотезы (Хи-квадрат)
```

```python
ct_o = df.groupby('Kidhome')['AcceptedCmp1'].value_counts().unstack()
ct_o
```

```python
scipy.stats.chi2_contingency(ct_o)
```

```python
# Доверительный интервал возьмем 99%, тогда 
# alpha = 0.01
# по таблице critical_value = 9.21034
# Следовательно, зависимость между категориями переменных есть, отвергаем H0, принимаем H1
```

#### Проверим существует ли статистически значимая корреляция между доходом респондентов и принятием решения о покупке в четвертую волну компании.

```python
# Выдвигаем гипотезы:
# H_1 : Между доходом респондентов и принятием решения о покупке в четвертую волну компании связь есть
# H_0 : Между доходом респондентов и принятием решения о покупке в четвертую волну компании связи нет

#Для поиска связи будем применять паратемтрический критерий проверки гипотезы (Т-тест)
```

```python
kupili = df[df['AcceptedCmp4'] == 1]['Income']
non_kupili = df[df['AcceptedCmp4'] == 0]['Income']
```

```python
## чтобы применить Т-тест нужно чтобы было равенство дисперсий и нормальное распределение
```

```python
kupili.hist()
```

```python
non_kupili.hist()
```

```python
# Видим, что распределение тяготеет к нормальному
```

```python
kupili.var()
```

```python
non_kupili.var()
```

```python
# так как дисперсии очень разные, поэтому используем не классический критерий Т Стьюдента,
# а рассчитаем его с поправкой Уэлча
```

```python
# высчитаем кол-во степеней свободы
dof = len(non_kupili)+ len(kupili) - 2
print (dof)
```

```python
n_1 = len(non_kupili)
n_2 = len(kupili)
print(n_1, n_2)
```

```python
var_1 = non_kupili.var()
var_2 = kupili.var()
print(var_1, var_2)
```

```python
v_1 = var_1/n_1
v_2 = var_2/n_2

df_welch = (v_1 + v_2)**2 / (v_1**2 / (n_1 - 1) + v_2**2 / (n_2 - 1))
print(df_welch)
print(int(df_welch))
```

```python
# Будем проверять на уровне значимости alpha = 0.05
# Так как dof = 2203
# По таблице распределения Стьюдента
t_critical = 1.9623
```

```python
def hypothesis_check(p_value, t_critical):
    if abs(p_value) < t_critical:
        print('Отвергаем Н0. Разница между средними статистически значима.')
    else:
        print('Не удалось отвергнуть H0')
```

```python
s_m1_sm2 = np.sqrt(v_1 +v_2)
print (s_m1_sm2)
```

```python
mean_1 = non_kupili.mean()
mean_2 = kupili.mean()
print(mean_1, mean_2)
```

```python
t = (mean_1 - mean_2) / s_m1_sm2
print(t)
```

```python
hypothesis_check(t, t_critical)
```

```python
#### Отвергаем H1, принимаем H0.
```

```python
# на всякий случай проведем анализ с помочью функции из библиотеки.
scipy.stats.ttest_ind_from_stats(mean_1, np.sqrt(var_1), n_1, 
                                 mean_2, np.sqrt(var_2), n_2, equal_var=False)
```

```python
# наличие линейной взаимосвязи больше всего визуально коррелирует между переменными потребления вина и потребления в целом
# однако при проверке на мультиколлениарность эти переменные оказались зависимыми (корреляция 0.9)
# поэтому возьмем другие переменные с более низкой кореляцией

X = df['MntMeatProducts']
y = df['MntTotal']

res = scipy.stats.linregress(X, y)
print(res)
```

```python
# выясним коэффициент детерминации, который показывает насколько хорошо модель описывает данные
r_sq = res.rvalue ** 2
print(r_sq)
```

```python
plt.plot(X, y, 'o', label='оригинальные данные')
plt.plot(X, res.intercept + res.slope*X, 'r', label='линия регрессии')
plt.legend()
plt.show()
```
